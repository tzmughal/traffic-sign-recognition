```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
import random

# Paths
dataset_dir = "dataset/GTSRB"
train_csv_path = os.path.join(dataset_dir, "Train.csv")
test_csv_path = os.path.join(dataset_dir, "Test.csv")
meta_csv_path = os.path.join(dataset_dir, "Meta.csv")

# Load CSVs
train_df = pd.read_csv(train_csv_path)
test_df = pd.read_csv(test_csv_path)
meta_df = pd.read_csv(meta_csv_path)





```

```python
# Show columns
print("Train.csv columns:", list(train_df.columns))
print("Test.csv columns:", list(test_df.columns))
print("Meta.csv columns:", list(meta_df.columns))
```

```python
# Show samples
print("\n--- Train.csv sample ---")
print(train_df.head())

print("\n--- Test.csv sample ---")
print(test_df.head())

print("\n--- Meta.csv sample ---")
print(meta_df.head())
```

```python

# Stats
print("\nNumber of training samples:", len(train_df))
print("Number of test samples:", len(test_df))
print("Number of unique classes:", train_df['ClassId'].nunique())
```

```python
# === Show random images from Train ===
def show_random_images(df, title, num_images=5):
    fig, axes = plt.subplots(1, num_images, figsize=(15, 3))
    fig.suptitle(title, fontsize=16)
    
    for ax in axes:
        sample = df.sample(1).iloc[0]
        img_path = os.path.join(dataset_dir, sample['Path'])
        img = Image.open(img_path)
        ax.imshow(img)
        ax.set_title(f"Class: {sample['ClassId']}")
        ax.axis("off")
    
    plt.show()

show_random_images(train_df, "Random Training Samples")
show_random_images(test_df, "Random Test Samples")
```
